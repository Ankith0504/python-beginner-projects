```python
#Reading image and printing it's dimensions

import cv2

image = cv2.imread('wolf.png')

x , y = image.shape[:2]

print("Height= {}, Width = {} ".format(x,y))
```

```python
#Selecting pixels of the certain location
(R, G, B) = image[120,70]

print("R = {}, G = {}, B = {}".format(R,G,B))

#Selecting only Blue
B = image[110,90,0]
print("It's a value of B = {}".format(B))
```

```python
#Resizing the image to any dimension
size = cv2.resize(image, (200,120))
cv2.imwrite("sizedimg.jpg", image)
```

```python
#Placing a box around any pixel
rectangle = cv2.rectangle(image, (250,170),(100,120),(122,0,0),2)
cv2.imwrite("newimg.jpg", image)
```

```python
#Adding text to the screen
text_input =cv2.putText(image , 'DONNN', (100,120), cv2.FONT_HERSHEY_SIMPLEX, 1, (122,0,0), 2)
cv2.imwrite("newimgg.jpg", image)
```
